```python
import trimesh
import numpy as np
```

```python
def load_mesh(path):

    scene_or_mesh = trimesh.load(path)

    # If file contains a single mesh:
    if isinstance(scene_or_mesh, trimesh.Trimesh):
        mesh = scene_or_mesh

    # If file contains a Scene (most .glb files do)
    else:
        scene = scene_or_mesh

        if not scene.geometry:
            raise ValueError("GLB file contains no mesh geometry.")

        # Extract first mesh from the scene
        mesh = list(scene.geometry.values())[0]
    
    v = mesh.vertices

    # Convert to tuples so numpy arrays become hashable
    as_tuples = [tuple(x) for x in v]

    from collections import Counter
    counts = Counter(as_tuples)

    duplicate_vertices = [v for v, c in counts.items() if c > 1]

    print("Number of unique vertices:", len(counts))
    print("Number of duplicates:", len(v) - len(counts))
    print("Duplicated entries (vertex coordinate):", duplicate_vertices)
    
    print("\nLoaded mesh:", path)
    print("Vertices:", len(mesh.vertices))
    print("Faces:", len(mesh.faces))
    print("Is indexed:", mesh.faces is not None)
    print("Is watertight:", mesh.is_watertight)
    print("Bounds:\n", mesh.bounds)
    print("Vertices:\n", mesh.vertices)

    return mesh
```
