```python
from plyfile import PlyData

# Load the PLY
ply_file = '/home/student/ss/data/scene_datasets/sim/stages/room_stage_semantic.ply'
plydata = PlyData.read(ply_file)

# Access vertices
vertices = plydata['vertex'].data

# Print the color of each vertex
for i, v in enumerate(vertices):
    r = v['red']
    g = v['green']
    b = v['blue']
    a = v['alpha'] if 'alpha' in v.dtype.names else 255
    print(f"Vertex {i}: R={r}, G={g}, B={b}, A={a}")
    print(v)
```

```python
from plyfile import PlyData, PlyElement
import numpy as np

plydata = PlyData.read(ply_file)

vertices = plydata['vertex'].data

# Create a copy of vertices with modified colors
new_vertices = []
for v in vertices:
    x, y, z = v['x'], v['y'], v['z']
    # floor semantic ID = 1
    new_vertices.append((x, y, z, 1, 0, 0, 255))  # R=1, G=0, B=0, A=255

# Define new dtype
vertex_dtype = [('x', 'f4'), ('y', 'f4'), ('z', 'f4'),
                ('red', 'u1'), ('green', 'u1'), ('blue', 'u1'), ('alpha', 'u1')]

vertex_array = np.array(new_vertices, dtype=vertex_dtype)
vertex_element = PlyElement.describe(vertex_array, 'vertex')

# Keep faces as is
face_element = plydata['face']

# Write new PLY
PlyData([vertex_element, face_element], text=False).write('/home/student/ss/data/scene_datasets/sim/stages/cube_stage_semantic_fixed.ply')
```
